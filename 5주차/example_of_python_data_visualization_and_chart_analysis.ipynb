# **Correlation**

- A **Scatter plot chart** will show the relationship between two different variables or it can reveal the distribution trends.

scatter_chart.PNG
- 연령별, 성별로 기록 관계를 알 수 있다.
- 여성보다는 남성의 기록이 더 좋은 것을 알 수 있고, 연령이 낮을수록 기록이 더 좋은 편이라는 것을 알 수 있다.

- A **bubble chart** is a variation of a scatter chart in which an additional data is represented in the size of the bubbles.

bubble_chart.PNG
- 마라톤 대회 도중 응급상황에 대처하기 위해 5대의 앰뷸런스를 배치하려 할 때 어느 지점에 배치하면 좋을 지 판단하는 근거가 될 수 있다.

- 2시간 대에 사람들이 어느 구간에 많이 있는지를 버블차트로 확인하여 앰뷸런스를 배치할 수 있다.

- A **Heat map** shows the relationship between two items and provides rating information, such as high to low or poor to excellent.

heat_map.PNG
- 어떤 연령대, 어떤 성별을 중심으로 마케팅을 해야하는지, 어떤 모델을 고용해야 하는지 알아보기 위함
- 연령을 세로축, 성별을 가로축으로 놓음
- 45세 남성, 여성의 비율이 가장 높기 때문에 해당 연령의 광고 모델을 고용해야 한다는 결론을 얻을 수 있다.
